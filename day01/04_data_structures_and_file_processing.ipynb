### ğŸ”§ ë¦¬ìŠ¤íŠ¸ ìœ ìš©í•œ ë©”ì„œë“œë“¤

ë¦¬ìŠ¤íŠ¸ë¥¼ ë” ì‰½ê²Œ ë‹¤ë£° ìˆ˜ ìˆëŠ” ë„êµ¬ë“¤ì„ ë°°ì›Œë´…ì‹œë‹¤!


```python
# ë¦¬ìŠ¤íŠ¸ ë©”ì„œë“œ ì‹¤ìŠµ (dir()ì™€ help() í™œìš©)
print("=== ë¦¬ìŠ¤íŠ¸ ë©”ì„œë“œ ë§ˆìŠ¤í„°í•˜ê¸° ===")

# ê³¼ì¼ ë¦¬ìŠ¤íŠ¸ë¡œ ì‹¤ìŠµ
fruits = ["ì‚¬ê³¼", "ë°”ë‚˜ë‚˜", "ì˜¤ë Œì§€"]
print(f"ì²˜ìŒ ê³¼ì¼ ëª©ë¡: {fruits}")

print("\nğŸ” ë¦¬ìŠ¤íŠ¸ ë©”ì„œë“œ íƒìƒ‰í•´ë³´ê¸°:")
print("dir(fruits)ë¡œ í™•ì¸í•œ ì£¼ìš” ë©”ì„œë“œë“¤:")
main_methods = ['append', 'insert', 'remove', 'pop', 'count', 'index', 'sort', 'reverse', 'clear', 'copy', 'extend']
print(main_methods)

print("\nğŸ“š ê° ë©”ì„œë“œ ì‹¤ìŠµ:")

# 1. append() - ëì— ì¶”ê°€
print("\n1. append() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.append(ê°’) - ë¦¬ìŠ¤íŠ¸ ëì— ê°’ì„ ì¶”ê°€")
fruits.append("í¬ë„")
print(f"   fruits.append('í¬ë„') â†’ {fruits}")

# 2. insert() - ì›í•˜ëŠ” ìœ„ì¹˜ì— ì¶”ê°€
print("\n2. insert() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.insert(ìœ„ì¹˜, ê°’) - íŠ¹ì • ìœ„ì¹˜ì— ê°’ì„ ì¶”ê°€")
fruits.insert(1, "ë”¸ê¸°")  # 1ë²ˆ ìœ„ì¹˜ì— ë”¸ê¸° ì¶”ê°€
print(f"   fruits.insert(1, 'ë”¸ê¸°') â†’ {fruits}")

# 3. remove() - íŠ¹ì • ê°’ ì‚­ì œ
print("\n3. remove() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.remove(ê°’) - ì²« ë²ˆì§¸ë¡œ ë‚˜ì˜¤ëŠ” ê°’ì„ ì‚­ì œ")
fruits.remove("ë°”ë‚˜ë‚˜")
print(f"   fruits.remove('ë°”ë‚˜ë‚˜') â†’ {fruits}")

# 4. pop() - ë§ˆì§€ë§‰ ìš”ì†Œ ì‚­ì œí•˜ê³  ë°˜í™˜
print("\n4. pop() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.pop() - ë§ˆì§€ë§‰ ìš”ì†Œë¥¼ ì‚­ì œí•˜ê³  ê·¸ ê°’ì„ ë°˜í™˜")
last_fruit = fruits.pop()
print(f"   last_fruit = fruits.pop() â†’ '{last_fruit}' ì œê±°, ë‚¨ì€ ê²ƒ: {fruits}")

# 5. count() - ê°’ì˜ ê°œìˆ˜ ì„¸ê¸°
print("\n5. count() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.count(ê°’) - íŠ¹ì • ê°’ì´ ëª‡ ê°œ ìˆëŠ”ì§€ ì„¸ê¸°")
fruits.extend(["ì‚¬ê³¼", "ì‚¬ê³¼"])  # ì‚¬ê³¼ 2ê°œ ë” ì¶”ê°€
apple_count = fruits.count("ì‚¬ê³¼")
print(f"   fruits.count('ì‚¬ê³¼') â†’ ì‚¬ê³¼ê°€ {apple_count}ê°œ ìˆì–´ìš”!")

# 6. index() - ê°’ì˜ ìœ„ì¹˜ ì°¾ê¸°
print("\n6. index() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.index(ê°’) - ê°’ì´ ìˆëŠ” ì²« ë²ˆì§¸ ìœ„ì¹˜ë¥¼ ë°˜í™˜")
strawberry_index = fruits.index("ë”¸ê¸°")
print(f"   fruits.index('ë”¸ê¸°') â†’ ë”¸ê¸°ëŠ” {strawberry_index}ë²ˆ ìœ„ì¹˜ì— ìˆì–´ìš”!")

# 7. sort() - ì •ë ¬í•˜ê¸°
print("\n7. sort() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.sort() - ë¦¬ìŠ¤íŠ¸ë¥¼ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬")
fruits.sort()
print(f"   fruits.sort() â†’ {fruits}")

# 8. reverse() - ìˆœì„œ ë’¤ì§‘ê¸°
print("\n8. reverse() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.reverse() - ë¦¬ìŠ¤íŠ¸ ìˆœì„œë¥¼ ë’¤ì§‘ê¸°")
fruits.reverse()
print(f"   fruits.reverse() â†’ {fruits}")

# 9. copy()ì™€ clear() - ë³µì‚¬ì™€ ë¹„ìš°ê¸°
print("\n9. copy()ì™€ clear() ë©”ì„œë“œ:")
print("   copy() ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.copy() - ë¦¬ìŠ¤íŠ¸ ë³µì‚¬ë³¸ ë§Œë“¤ê¸°")
print("   clear() ì‚¬ìš©ë²•: ë¦¬ìŠ¤íŠ¸.clear() - ë¦¬ìŠ¤íŠ¸ì˜ ëª¨ë“  ìš”ì†Œ ì‚­ì œ")
fruits_copy = fruits.copy()  # ë³µì‚¬ë³¸ ë§Œë“¤ê¸°
fruits_copy.clear()
print(f"   ë³µì‚¬ë³¸ ë¹„ìš°ê¸°: {fruits_copy}")
print(f"   ì›ë³¸ì€ ê·¸ëŒ€ë¡œ: {fruits}")

print("\nğŸ’¡ ë©”ì„œë“œ ì‚¬ìš© íŒ:")
print("â€¢ ê¶ê¸ˆí•œ ë©”ì„œë“œê°€ ìˆìœ¼ë©´ help(list.ë©”ì„œë“œëª…)ìœ¼ë¡œ ìì„¸í•œ ì„¤ëª…ì„ ë³¼ ìˆ˜ ìˆì–´ìš”!")
print("â€¢ ì˜ˆ: help(list.append), help(list.sort)")
print("â€¢ ì‹¤ì œë¡œ ì‚¬ìš©í•´ë³´ë©´ì„œ ìµíˆëŠ” ê²ƒì´ ê°€ì¥ ì¢‹ì•„ìš”!")
```

```python
# ë”•ì…”ë„ˆë¦¬ ë©”ì„œë“œ ì‹¤ìŠµ (dir()ì™€ help() í™œìš©)
print("=== ë”•ì…”ë„ˆë¦¬ ë©”ì„œë“œ ë§ˆìŠ¤í„°í•˜ê¸° ===")

# í•™ìƒ ì ìˆ˜ ë”•ì…”ë„ˆë¦¬ë¡œ ì‹¤ìŠµ
scores = {"ì² ìˆ˜": 85, "ì˜í¬": 92, "ë¯¼ìˆ˜": 78}
print(f"ì²˜ìŒ ì ìˆ˜: {scores}")

print("\nğŸ” ë”•ì…”ë„ˆë¦¬ ë©”ì„œë“œ íƒìƒ‰í•´ë³´ê¸°:")
print("dir(scores)ë¡œ í™•ì¸í•œ ì£¼ìš” ë©”ì„œë“œë“¤:")
main_methods = ['get', 'keys', 'values', 'items', 'update', 'pop', 'popitem', 'clear', 'copy', 'setdefault']
print(main_methods)

print("\nğŸ“š ê° ë©”ì„œë“œ ì‹¤ìŠµ:")

# 1. get() - ì•ˆì „í•˜ê²Œ ê°’ ê°€ì ¸ì˜¤ê¸°
print("\n1. get() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.get(í‚¤, ê¸°ë³¸ê°’) - í‚¤ê°€ ì—†ì–´ë„ ì—ëŸ¬ ì•ˆë‚¨")
ì² ìˆ˜_ì ìˆ˜ = scores.get("ì² ìˆ˜")
ì§€ì—°_ì ìˆ˜ = scores.get("ì§€ì—°", 0)  # ì—†ìœ¼ë©´ ê¸°ë³¸ê°’ 0
print(f"   scores.get('ì² ìˆ˜') â†’ {ì² ìˆ˜_ì ìˆ˜}")
print(f"   scores.get('ì§€ì—°', 0) â†’ {ì§€ì—°_ì ìˆ˜} (ì—†ì–´ì„œ ê¸°ë³¸ê°’)")

# 2. keys() - ëª¨ë“  í‚¤ ê°€ì ¸ì˜¤ê¸°
print("\n2. keys() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.keys() - ëª¨ë“  í‚¤ë¥¼ ê°€ì ¸ì˜´")
student_names = list(scores.keys())
print(f"   list(scores.keys()) â†’ {student_names}")

# 3. values() - ëª¨ë“  ê°’ ê°€ì ¸ì˜¤ê¸°
print("\n3. values() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.values() - ëª¨ë“  ê°’ì„ ê°€ì ¸ì˜´")
all_scores = list(scores.values())
print(f"   list(scores.values()) â†’ {all_scores}")

# 4. items() - í‚¤ì™€ ê°’ì„ ìŒìœ¼ë¡œ ê°€ì ¸ì˜¤ê¸°
print("\n4. items() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.items() - (í‚¤, ê°’) ìŒì„ ê°€ì ¸ì˜´")
print("   for ë°˜ë³µë¬¸ì—ì„œ ìì£¼ ì‚¬ìš©:")
for name, score in scores.items():
    print(f"     {name}: {score}ì ")

# 5. update() - ì—¬ëŸ¬ í•­ëª© í•œë²ˆì— ì¶”ê°€/ìˆ˜ì •
print("\n5. update() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.update(ë‹¤ë¥¸ë”•ì…”ë„ˆë¦¬) - ì—¬ëŸ¬ í•­ëª© í•œë²ˆì— ì¶”ê°€/ìˆ˜ì •")
new_scores = {"ì§€ì—°": 88, "ë‹¤ìš´": 95, "ì² ìˆ˜": 90}  # ì² ìˆ˜ëŠ” ì ìˆ˜ ìˆ˜ì •
scores.update(new_scores)
print(f"   scores.update({new_scores}) â†’ {scores}")

# 6. pop() - í‚¤ë¡œ ì‚­ì œí•˜ê³  ê°’ ë°˜í™˜
print("\n6. pop() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.pop(í‚¤) - í‚¤ë¥¼ ì‚­ì œí•˜ê³  ê·¸ ê°’ì„ ë°˜í™˜")
ë¯¼ìˆ˜_ì ìˆ˜ = scores.pop("ë¯¼ìˆ˜")
print(f"   scores.pop('ë¯¼ìˆ˜') â†’ {ë¯¼ìˆ˜_ì ìˆ˜} ë°˜í™˜, ë‚¨ì€ ê²ƒ: {scores}")

# 7. popitem() - ë§ˆì§€ë§‰ í•­ëª© ì‚­ì œí•˜ê³  ë°˜í™˜
print("\n7. popitem() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.popitem() - ë§ˆì§€ë§‰ (í‚¤, ê°’) ìŒì„ ì‚­ì œí•˜ê³  ë°˜í™˜")
last_item = scores.popitem()
print(f"   scores.popitem() â†’ {last_item} ë°˜í™˜, ë‚¨ì€ ê²ƒ: {scores}")

# 8. copy()ì™€ clear() - ë³µì‚¬ì™€ ë¹„ìš°ê¸°
print("\n8. copy()ì™€ clear() ë©”ì„œë“œ:")
print("   copy() ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.copy() - ë”•ì…”ë„ˆë¦¬ ë³µì‚¬ë³¸ ë§Œë“¤ê¸°")
print("   clear() ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.clear() - ëª¨ë“  í•­ëª© ì‚­ì œ")
scores_copy = scores.copy()  # ë³µì‚¬ë³¸ ë§Œë“¤ê¸°
scores_copy.clear()
print(f"   ë³µì‚¬ë³¸ ë¹„ìš°ê¸°: {scores_copy}")
print(f"   ì›ë³¸ì€ ê·¸ëŒ€ë¡œ: {scores}")

# 9. setdefault() - í‚¤ê°€ ì—†ìœ¼ë©´ ê¸°ë³¸ê°’ìœ¼ë¡œ ì¶”ê°€
print("\n9. setdefault() ë©”ì„œë“œ:")
print("   ì‚¬ìš©ë²•: ë”•ì…”ë„ˆë¦¬.setdefault(í‚¤, ê¸°ë³¸ê°’) - í‚¤ê°€ ì—†ìœ¼ë©´ ê¸°ë³¸ê°’ìœ¼ë¡œ ì¶”ê°€")
scores.setdefault("í•˜ëŠ˜", 80)  # í•˜ëŠ˜ì´ ì—†ìœ¼ë©´ 80ì ìœ¼ë¡œ ì¶”ê°€
scores.setdefault("ì² ìˆ˜", 60)   # ì² ìˆ˜ëŠ” ì´ë¯¸ ìˆìœ¼ë‹ˆ ë³€ê²½ ì•ˆë¨
print(f"   scores.setdefault('í•˜ëŠ˜', 80) â†’ {scores}")
print(f"   scores.setdefault('ì² ìˆ˜', 60) â†’ ì² ìˆ˜ëŠ” ì´ë¯¸ ìˆì–´ì„œ ë³€ê²½ ì•ˆë¨")

print("\nğŸ’¡ ë”•ì…”ë„ˆë¦¬ ë©”ì„œë“œ ì‚¬ìš© íŒ:")
print("â€¢ get() vs [í‚¤]: get()ì€ í‚¤ê°€ ì—†ì–´ë„ ì—ëŸ¬ ì•ˆë‚¨, []ëŠ” ì—ëŸ¬ ë°œìƒ")
print("â€¢ items()ëŠ” for ë°˜ë³µë¬¸ì—ì„œ í‚¤ì™€ ê°’ì„ ë™ì‹œì— ì‚¬ìš©í•  ë•Œ ìœ ìš©")
print("â€¢ update()ëŠ” ì—¬ëŸ¬ ë°ì´í„°ë¥¼ í•œë²ˆì— ì¶”ê°€í•  ë•Œ í¸ë¦¬")
print("â€¢ help(dict.get), help(dict.update) ë“±ìœ¼ë¡œ ìì„¸í•œ ì‚¬ìš©ë²• í™•ì¸ ê°€ëŠ¥!")
```

## 2.6 ì§‘í•©(Set) - ì¤‘ë³µ ì—†ëŠ” ë°ì´í„° ëª¨ìŒ ğŸ¯

ì§‘í•©ì€ ì¤‘ë³µëœ ê°’ì„ í—ˆìš©í•˜ì§€ ì•ŠëŠ” íŠ¹ë³„í•œ ìë£Œêµ¬ì¡°ì…ë‹ˆë‹¤!


```python
# ì§‘í•© ê¸°ì´ˆì™€ ë©”ì„œë“œ
print("=== ì§‘í•© ì•Œì•„ë³´ê¸° ===")

# ì§‘í•© ë§Œë“¤ê¸° - ì¤‘ê´„í˜¸{} ì‚¬ìš©í•˜ê±°ë‚˜ set() í•¨ìˆ˜
fruits_set = {"ì‚¬ê³¼", "ë°”ë‚˜ë‚˜", "ì˜¤ë Œì§€", "ì‚¬ê³¼"}  # ì‚¬ê³¼ê°€ ì¤‘ë³µë¨
print(f"ê³¼ì¼ ì§‘í•©: {fruits_set}")  # ì¤‘ë³µ ì œê±°ë¨!
print(f"íƒ€ì…: {type(fruits_set)}")

# ë¦¬ìŠ¤íŠ¸ì—ì„œ ì§‘í•©ìœ¼ë¡œ ë³€í™˜ (ì¤‘ë³µ ì œê±°í•  ë•Œ ìœ ìš©!)
numbers_list = [1, 2, 2, 3, 3, 3, 4, 5]
numbers_set = set(numbers_list)
print(f"\nì›ë˜ ë¦¬ìŠ¤íŠ¸: {numbers_list}")
print(f"ì§‘í•©ìœ¼ë¡œ ë³€í™˜: {numbers_set}")

# ì§‘í•© ë©”ì„œë“œë“¤
animals = {"ê°•ì•„ì§€", "ê³ ì–‘ì´", "í† ë¼"}
print(f"\në™ë¬¼ ì§‘í•©: {animals}")

# 1. add() - ìš”ì†Œ í•˜ë‚˜ ì¶”ê°€
animals.add("í–„ìŠ¤í„°")
print(f"í–„ìŠ¤í„° ì¶”ê°€ í›„: {animals}")

# 2. update() - ì—¬ëŸ¬ ìš”ì†Œ í•œë²ˆì— ì¶”ê°€
animals.update(["ê±°ë¶ì´", "ê¸ˆë¶•ì–´"])
print(f"ì—¬ëŸ¬ ë™ë¬¼ ì¶”ê°€ í›„: {animals}")

# 3. remove() - ìš”ì†Œ ì‚­ì œ (ì—†ìœ¼ë©´ ì—ëŸ¬)
animals.remove("í† ë¼")
print(f"í† ë¼ ì œê±° í›„: {animals}")

# 4. discard() - ìš”ì†Œ ì‚­ì œ (ì—†ì–´ë„ ì—ëŸ¬ ì•ˆë‚¨)
animals.discard("ì‚¬ì")  # ì—†ì§€ë§Œ ì—ëŸ¬ ì•ˆë‚¨
animals.discard("ê³ ì–‘ì´")  # ìˆìœ¼ë©´ ì‚­ì œ
print(f"discard í›„: {animals}")

# 5. pop() - ì„ì˜ì˜ ìš”ì†Œ ì‚­ì œí•˜ê³  ë°˜í™˜
removed_animal = animals.pop()
print(f"'{removed_animal}' ì œê±° í›„: {animals}")

# 6. clear() - ëª¨ë“  ìš”ì†Œ ì‚­ì œ
animals_copy = animals.copy()
animals_copy.clear()
print(f"ë³µì‚¬ë³¸ ë¹„ìš°ê¸°: {animals_copy}")

# ì§‘í•©ì˜ íŠ¹ë³„í•œ ì—°ì‚°ë“¤
print(f"\n=== ì§‘í•© ì—°ì‚° ===")
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print(f"ì§‘í•© A: {A}")
print(f"ì§‘í•© B: {B}")

# 7. union() ë˜ëŠ” | - í•©ì§‘í•©
union_result = A.union(B)
print(f"í•©ì§‘í•© A âˆª B: {union_result}")

# 8. intersection() ë˜ëŠ” & - êµì§‘í•©
intersection_result = A.intersection(B)
print(f"êµì§‘í•© A âˆ© B: {intersection_result}")

# 9. difference() ë˜ëŠ” - - ì°¨ì§‘í•©
difference_result = A.difference(B)
print(f"ì°¨ì§‘í•© A - B: {difference_result}")

# 10. symmetric_difference() ë˜ëŠ” ^ - ëŒ€ì¹­ì°¨ì§‘í•©
symmetric_diff = A.symmetric_difference(B)
print(f"ëŒ€ì¹­ì°¨ì§‘í•© A â–³ B: {symmetric_diff}")

# ì§‘í•©ì´ ìœ ìš©í•œ ì‹¤ì œ ì˜ˆì‹œ
print(f"\n=== ì§‘í•©ì˜ ì‹¤ìš©ì  í™œìš© ===")

# ì˜ˆì‹œ 1: ì¤‘ë³µ ì œê±°
student_votes = ["ì² ìˆ˜", "ì˜í¬", "ì² ìˆ˜", "ë¯¼ìˆ˜", "ì˜í¬", "ì² ìˆ˜"]
unique_candidates = set(student_votes)
print(f"íˆ¬í‘œ ê²°ê³¼: {student_votes}")
print(f"í›„ë³´ìë“¤: {unique_candidates}")

# ì˜ˆì‹œ 2: ê³µí†µ ì·¨ë¯¸ ì°¾ê¸°
alice_hobbies = {"ë…ì„œ", "ì˜í™”ê°ìƒ", "ìš”ë¦¬", "ë“±ì‚°"}
bob_hobbies = {"ì˜í™”ê°ìƒ", "ê²Œì„", "ë“±ì‚°", "ìŒì•…ê°ìƒ"}
common_hobbies = alice_hobbies & bob_hobbies
print(f"\nAlice ì·¨ë¯¸: {alice_hobbies}")
print(f"Bob ì·¨ë¯¸: {bob_hobbies}")
print(f"ê³µí†µ ì·¨ë¯¸: {common_hobbies}")

# ì˜ˆì‹œ 3: ë©¤ë²„ì‹­ í™•ì¸
vip_members = {"ê¹€ì² ìˆ˜", "ì´ì˜í¬", "ë°•ë¯¼ìˆ˜"}
current_user = "ê¹€ì² ìˆ˜"
if current_user in vip_members:
    print(f"\nğŸŒŸ {current_user}ë‹˜ì€ VIP íšŒì›ì…ë‹ˆë‹¤!")
else:
    print(f"\n{current_user}ë‹˜ì€ ì¼ë°˜ íšŒì›ì…ë‹ˆë‹¤.")
```
