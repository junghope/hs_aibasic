### 2.2 while ë°˜ë³µë¬¸


````python
# while ë°˜ë³µë¬¸ ê¸°ë³¸ ì˜ˆì‹œ
count = 1
while count <= 5:
    print(f"ì¹´ìš´íŠ¸: {count}")
    count += 1

# breakì™€ continue í™œìš©
print("1ë¶€í„° 10ê¹Œì§€ ìˆ«ì ì¤‘ 3ì˜ ë°°ìˆ˜ê°€ ì•„ë‹Œ ìˆ˜:")
for i in range(1, 11):
    if i % 3 == 0:
        continue  # 3ì˜ ë°°ìˆ˜ëŠ” ê±´ë„ˆë›°ê¸°
    if i > 8:
        break  # 8ë³´ë‹¤ í¬ë©´ ë°˜ë³µ ì¤‘ë‹¨
    print(i, end=" ")
print()

# ë¬´í•œ ë°˜ë³µë¬¸ (while True)
print("\n=== ë¬´í•œ ë°˜ë³µë¬¸ ì˜ˆì‹œ ===")
menu_choices = ["1", "2", "3"]  # ì‹œë®¬ë ˆì´ì…˜ìš© ì„ íƒì§€
choice_index = 0

while True:
    print("\n--- ê°„ë‹¨í•œ ê³„ì‚°ê¸° ---")
    print("1. ë§ì…ˆ")
    print("2. ëº„ì…ˆ")
    print("3. ì¢…ë£Œ")

    # ì‹œë®¬ë ˆì´ì…˜ì„ ìœ„í•œ ì„ íƒ (ì‹¤ì œë¡œëŠ” input() ì‚¬ìš©)
    if choice_index < len(menu_choices):
        choice = menu_choices[choice_index]
        print(f"ì„ íƒ: {choice}")
        choice_index += 1
    else:
        choice = "3"

    if choice == "1":
        print("5 + 3 = 8")
    elif choice == "2":
        print("5 - 3 = 2")
    elif choice == "3":
        print("ê³„ì‚°ê¸°ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        break  # ë¬´í•œ ë°˜ë³µë¬¸ íƒˆì¶œ
    else:
        print("ì˜¬ë°”ë¥¸ ë²ˆí˜¸ë¥¼ ì„ íƒí•´ì£¼ì„¸ìš”.")

    # ì‹œë®¬ë ˆì´ì…˜ ì¢…ë£Œ ì¡°ê±´
    if choice_index >= len(menu_choices):
        break
````

## 3. í•¨ìˆ˜ ì •ì˜ì™€ í™œìš©

### 3.1 ê¸°ë³¸ í•¨ìˆ˜ ì •ì˜


````python
# ê¸°ë³¸ í•¨ìˆ˜ ì˜ˆì œ

# 1. í•™ì  ê³„ì‚° í•¨ìˆ˜
def calculate_grade(score):
    """ì ìˆ˜ë¥¼ ë°›ì•„ í•™ì ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F"

# 2. ê°„ë‹¨í•œ ê³„ì‚° í•¨ìˆ˜
def add_numbers(a, b):
    """ë‘ ìˆ˜ë¥¼ ë”í•˜ëŠ” í•¨ìˆ˜"""
    return a + b

# í•¨ìˆ˜ ì‚¬ìš© ì˜ˆì‹œ
print("=== í•¨ìˆ˜ ì‚¬ìš© ì˜ˆì‹œ ===")

# í•™ì  ê³„ì‚°
score = 85
grade = calculate_grade(score)
print(f"ì ìˆ˜ {score}: {grade}í•™ì ")

# ë§ì…ˆ í•¨ìˆ˜
result = add_numbers(10, 20)
print(f"10 + 20 = {result}")

# 3. ì¬ê·€í•¨ìˆ˜ (Recursion)
print("\n=== ì¬ê·€í•¨ìˆ˜ ì˜ˆì‹œ ===")

def factorial(n):
    """íŒ©í† ë¦¬ì–¼ì„ ê³„ì‚°í•˜ëŠ” ì¬ê·€í•¨ìˆ˜"""
    if n <= 1:  # ê¸°ì € ì¡°ê±´ (Base case)
        return 1
    else:
        return n * factorial(n - 1)  # ì¬ê·€ í˜¸ì¶œ

def countdown(n):
    """ì¹´ìš´íŠ¸ë‹¤ìš´í•˜ëŠ” ì¬ê·€í•¨ìˆ˜"""
    if n <= 0:  # ê¸°ì € ì¡°ê±´
        print("ë°œì‚¬!")
    else:
        print(n)
        countdown(n - 1)  # ì¬ê·€ í˜¸ì¶œ

# ì¬ê·€í•¨ìˆ˜ ì‚¬ìš© ì˜ˆì‹œ
print("5! =", factorial(5))  # 5! = 5 Ã— 4 Ã— 3 Ã— 2 Ã— 1 = 120
print("\nì¹´ìš´íŠ¸ë‹¤ìš´:")
countdown(3)

# 4. ëŒë‹¤ í•¨ìˆ˜ ê¸°ë³¸ ì˜ˆì‹œ
print("\n=== ëŒë‹¤ í•¨ìˆ˜ ì˜ˆì‹œ ===")
numbers = [1, 2, 3, 4, 5]

# filter: ì§ìˆ˜ë§Œ í•„í„°ë§
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"ì§ìˆ˜: {even_numbers}")

# map: ê° ìš”ì†Œë¥¼ ì œê³±
squared = list(map(lambda x: x**2, numbers))
print(f"ì œê³±: {squared}")
````
