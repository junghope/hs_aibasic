## 1.🤖 머신러닝이란?
- 명시적으로 프로그래밍하지 않고도 컴퓨터가 학습할 수 있게 하는 연구 분야
- 데이터로부터 패턴을 찾아 예측이나 결정을 내리는 알고리즘


### 1.1 🤖 전통적 프로그래밍 vs 머신러닝

```
전통적 프로그래밍:
데이터 + 프로그램 → 결과

머신러닝:
데이터 + 결과 → 프로그램(모델)
```

**💡 머신러닝이 유용한 상황들:**
1. **규칙이 복잡하거나 계속 변할 때**
   - 스팸 메일 분류 (새로운 스팸 기법 등장)
   - 주식 가격 예측 (시장 상황 변화)

2. **사람이 직접 규칙을 만들기 어려울 때**
   - 이미지 인식 (고양이 vs 강아지)
   - 음성 인식 (다양한 억양과 발음)

3. **대량의 데이터에서 패턴을 찾아야 할 때**
   - 고객 구매 패턴 분석
   - 의료 진단 보조

### 1.2 지도학습 vs 비지도학습 완전 이해

**🎯 지도학습 (Supervised Learning)**

**정의**: 정답(라벨)이 있는 데이터로 학습하여 새로운 데이터의 정답을 예측

**언제 사용?**
- ✅ 명확한 목표값(타겟)이 있을 때
- ✅ 과거 데이터와 결과가 충분할 때
- ✅ 예측이나 분류가 목적일 때

**대표 알고리즘:**
- 분류(Classification): 데이터가 범주형인 변수를 예측
- 회귀(Regression):  데이터가 연속인 값을 예측

**실생활 예시:**
- 📧 **스팸 메일 분류**: 과거 스팸/정상 메일 데이터로 학습
- 🏠 **집값 예측**: 지역, 면적, 방 개수 → 가격
- 🎬 **영화 추천**: 사용자 평점 데이터 → 선호도 예측
- 🏥 **질병 진단**: 검사 결과 → 질병 여부



**🔍 비지도학습 (Unsupervised Learning)**

**정의**: 정답 없이 데이터의 숨겨진 패턴이나 구조를 발견

**언제 사용?**
- ✅ 데이터의 패턴을 탐색하고 싶을 때
- ✅ 고객 세분화가 필요할 때
- ✅ 이상치를 찾고 싶을 때

**대표 알고리즘:**
- 클러스터링(Clustering): 특정 기준에 따라 유사한 데이터 사례들을 하나의 세트로 그룹화
- 차원 축소(Dimension Reduction): 고려 중인 변수의 개수를 줄이는 작업

**실생활 예시:**
- 👥 **고객 세분화**: 구매 패턴으로 고객 그룹 나누기
- 🛍️ **상품 연관성**: 함께 구매되는 상품 찾기
- 🔍 **이상 거래 탐지**: 평소와 다른 거래 패턴 발견
- 📰 **토픽 모델링**: 뉴스 기사의 주제 자동 분류

## 2. 회귀와 분류 문제

### 2.1 회귀 문제 (Regression)

**🎯 회귀 문제란?**
연속적인 숫자 값을 예측하는 문제입니다.

**📊 회귀 문제 특징:**
- 예측값이 **연속적인 수치**
- 중간값들도 의미가 있음
- 예측 정확도는 **오차의 크기**로 측정

**💼 회귀 문제 실생활 예시:**

| 분야 | 예측 목표 | 입력 변수 | 활용 |
|------|-----------|-----------|------|
| **부동산** | 집값 | 면적, 위치, 년수 | 시세 책정 |
| **마케팅** | 매출액 | 광고비, 계절성 | 예산 계획 |
| **제조** | 생산량 | 원자재, 인력 | 생산 계획 |
| **금융** | 주가 | 경제지표, 뉴스 | 투자 결정 |
| **의료** | 혈압 | 나이, 체중, 운동 | 건강 관리 |

### 2.2 분류 문제 (Classification)

**🎯 분류 문제란?**
미리 정해진 범주(클래스) 중 하나를 선택하는 문제입니다.

**📋 분류 문제 특징:**
- 예측값이 **범주형 데이터**
- 이진 분류(2개) vs 다중 분류(3개 이상)
- 예측 정확도는 **정답률**로 측정

**💼 분류 문제 실생활 예시:**

| 분야 | 분류 목표 | 클래스 | 활용 |
|------|-----------|--------|------|
| **IT** | 스팸 메일 | 스팸/정상 | 메일 필터링 |
| **의료** | 질병 진단 | 양성/음성 | 조기 진단 |
| **금융** | 대출 승인 | 승인/거절 | 리스크 관리 |
| **교육** | 학습 상태 | 우수/보통/부족 | 맞춤 교육 |
| **이커머스** | 고객 등급 | VIP/일반/신규 | 차별화 서비스 |

### 2.3 회귀 vs 분류 구분 연습

**🤔 다음 중 회귀 문제는?**

1. ✅ **온도 예측** (25.3도, 32.7도) → 회귀
2. ❌ **날씨 예측** (맑음, 흐림, 비) → 분류
3. ✅ **시험 점수 예측** (87점, 92점) → 회귀
4. ❌ **합격 여부 예측** (합격, 불합격) → 분류
5. ✅ **주식 가격 예측** (50,000원) → 회귀
6. ❌ **주식 추천** (매수, 보유, 매도) → 분류

````python
# 🌟 초보자를 위한 간단한 머신러닝 예제
# 분류: 붓꽃 종류 맞추기
# 회귀: 집 크기로 가격 맞추기

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score
import numpy as np

# ==========================================
# 🌸 Part 1: 분류 - 붓꽃 종류 맞추기
# ==========================================
print("\n📚 Part 1: 분류 문제")
print("목표: 꽃의 크기로 붓꽃 종류를 맞춰보자!")

# 1단계: 데이터 가져오기
print("\n1단계: 데이터 준비")
iris = load_iris()
flower_data = iris.data          # 꽃의 크기 정보
flower_type = iris.target        # 정답 (어떤 종류인지)

print(f"꽃 데이터 개수: {len(flower_data)}개")
print(f"꽃 종류: {iris.target_names}")  # ['setosa' 'versicolor' 'virginica']

# 2단계: 데이터 살펴보기
print("\n2단계: 데이터 미리보기")
# 처음 5개 꽃의 정보
print("처음 5개 꽃의 크기:")
for i in range(5):
    print(f"꽃 {i+1}: {flower_data[i]} → 종류: {iris.target_names[flower_type[i]]}")

# 3단계: 훈련용과 테스트용으로 나누기
print("\n3단계: 데이터 나누기")
X_train, X_test, y_train, y_test = train_test_split(
    flower_data, flower_type, test_size=0.3, random_state=42)

print(f"훈련용 데이터: {len(X_train)}개")
print(f"테스트용 데이터: {len(X_test)}개")

# 4단계: AI 모델 만들기
print("\n4단계: AI 모델 훈련")
classification_model = LogisticRegression()
classification_model.fit(X_train, y_train)  # 훈련하기
print("✅ 모델 훈련 완료!")

# 5단계: 예측해보기
print("\n5단계: 예측 결과")
predictions = classification_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"정확도: {accuracy:.2f} ({accuracy*100:.1f}%)")

# 몇 개 예측 결과 보여주기
print("\n예측 결과 예시:")
for i in range(5):
    actual = iris.target_names[y_test[i]]
    predicted = iris.target_names[predictions[i]]
    correct = "O" if actual == predicted else "X"
    print(f"실제: {actual:10} | 예측: {predicted:10} {correct}")




````

✅ 분류 (Classification)
  - 목표: 카테고리를 맞추기 (붓꽃 종류)
  - 결과: 정확도로 평가
  - 예시: 스팸 메일 분류, 동물 사진 분류

✅ 회귀 (Regression)  
  - 목표: 숫자를 예측하기 (집 가격)
  - 결과: 실제값과 예측값 비교
  - 예시: 주식 가격 예측, 기온 예측

💡 머신러닝 기본 단계:
  - 1️⃣ 데이터 준비
  - 2️⃣ 훈련용/테스트용 나누기
  - 3️⃣ 모델 훈련
  - 4️⃣ 예측하기
  - 5️⃣ 결과 평가



## 3. 모델 평가 지표 

### 3.1 회귀 모델 평가 지표

**📏 평균 절대 오차 (MAE - Mean Absolute Error)**
```
MAE = |실제값 - 예측값|의 평균
```
- **장점**: 직관적이고 이해하기 쉬움
- **단점**: 큰 오차에 덜 민감
- **예시**: 집값 예측에서 평균 1000만원 차이

**📐 평균 제곱 오차 (MSE - Mean Square Error)**
```
MSE = (실제값 - 예측값)²의 평균
```
- **장점**: 큰 오차에 더 큰 패널티
- **단점**: 단위가 제곱되어 해석 어려움
- **예시**: 온도 예측에서 MSE = 4 (평균 2도 차이)

**📊 결정계수 (R²)**
```
R² = 1 - (예측 오차 / 전체 변동)
```
- **범위**: 0~1 (1에 가까울수록 좋음)
- **의미**: 모델이 설명하는 변동의 비율
- **예시**: R² = 0.8 → 80% 설명력

### 3.3 분류 모델 평가 지표

**🎯 정확도 (Accuracy)**
```
정확도 = 맞춘 개수 / 전체 개수
```
- **장점**: 가장 직관적
- **단점**: 불균형 데이터에서 오해의 소지
- **예시**: 100명 중 85명 맞춤 = 85%

**🔍 정밀도 (Precision)**
```
정밀도 = 진짜 양성 / (진짜 양성 + 가짜 양성)
```
- **의미**: 양성으로 예측한 것 중 실제 양성 비율
- **중요할 때**: 가짜 알람이 치명적인 경우
- **예시**: 스팸 분류 - 정상 메일을 스팸으로 분류하면 안 됨

**📡 재현율 (Recall)**
```
재현율 = 진짜 양성 / (진짜 양성 + 가짜 음성)
```
- **의미**: 실제 양성 중 찾아낸 비율
- **중요할 때**: 놓치면 안 되는 경우
- **예시**: 질병 진단 - 환자를 놓치면 안 됨

## 4. 과적합과 일반화 완전 이해

### 4.1 과적합 (Overfitting) 문제

**🎯 과적합이란?**
모델이 **훈련 데이터에만 너무 특화**되어 새로운 데이터에서 성능이 떨어지는 현상

<img src='https://drive.google.com/uc?export=download&id=1xGfanYC3FlEe7M93Qn3Ip48jgZLGYS80' width="739" height ="296" /><br>

##### 출처: https://kmina02.tistory.com/167

**🔍 과적합 징후:**
- 훈련 데이터 정확도: 99%
- 테스트 데이터 정확도: 70%
- → 29%의 큰 차이 발생!

**💡 과적합 원인:**
1. **모델이 너무 복잡**함 (파라미터가 많음)
2. **훈련 데이터가 부족**함
3. **노이즈가 많은 데이터**
4. **훈련을 너무 오래**함

**🛠️ 과적합 해결 방법:**

| 방법 | 설명 | 예시 |
|------|------|------|
| **정규화** | 모델 복잡도에 패널티 | L1, L2 정규화 |
| **조기 종료** | 검증 성능이 떨어지면 중단 | Early Stopping |
| **데이터 증강** | 더 많은 훈련 데이터 확보 | 회전, 크기 변경 |
| **교차 검증** | 여러 번 나누어 검증 | k-fold CV |
| **드롭아웃** | 일부 뉴런을 무작위로 제거 | 신경망에서 사용 |

````python
# 모델 평가 지표 실습
print("\\n=== 모델 평가 지표 실습 ===")

# 회귀 평가 지표 예제
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 간단한 회귀 데이터
X_simple = np.array([[1], [2], [3], [4], [5]])
y_true = np.array([2, 4, 6, 8, 10])
y_pred = np.array([2.1, 3.9, 6.2, 7.8, 9.9])

print("회귀 평가 지표:")
print(f"MAE (평균 절대 오차): {mean_absolute_error(y_true, y_pred):.3f}")
print(f"MSE (평균 제곱 오차): {mean_squared_error(y_true, y_pred):.3f}")
print(f"R² (결정계수): {r2_score(y_true, y_pred):.3f}")

# 분류 평가 지표 예제
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# 분류 예제 데이터 (스팸 메일 분류)
y_true_class = np.array([1, 0, 1, 1, 0, 1, 0, 0, 1, 0])  # 실제 (1: 스팸, 0: 정상)
y_pred_class = np.array([1, 0, 0, 1, 0, 1, 0, 1, 1, 0])  # 예측

print("\\n분류 평가 지표:")
print(f"정확도 (Accuracy): {accuracy_score(y_true_class, y_pred_class):.3f}")
print(f"정밀도 (Precision): {precision_score(y_true_class, y_pred_class):.3f}")
print(f"재현율 (Recall): {recall_score(y_true_class, y_pred_class):.3f}")

print("\\n혼동 행렬:")
cm = confusion_matrix(y_true_class, y_pred_class)
print(f"실제 정상, 예측 정상 (TN): {cm[0,0]}")
print(f"실제 정상, 예측 스팸 (FP): {cm[0,1]}")
print(f"실제 스팸, 예측 정상 (FN): {cm[1,0]}")
print(f"실제 스팸, 예측 스팸 (TP): {cm[1,1]}")

# 과적합 시각화
print("\\n=== 과적합 vs 일반화 시각화 ===")

# 다항식 회귀로 과적합 시연
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# 실제 데이터 생성 (sin 함수 + 노이즈)
np.random.seed(42)
X_demo = np.linspace(0, 1, 20).reshape(-1, 1)
y_demo = np.sin(2 * np.pi * X_demo).ravel() + np.random.normal(0, 0.3, X_demo.shape[0])

# 테스트 데이터
X_test_demo = np.linspace(0, 1, 100).reshape(-1, 1)
y_test_true = np.sin(2 * np.pi * X_test_demo).ravel()

plt.figure(figsize=(15, 5))

degrees = [1, 4, 15]  # 단순한 모델, 적절한 모델, 과적합 모델
titles = ['과소적합 (Underfitting)', '적절한 모델', '과적합 (Overfitting)']

for i, (degree, title) in enumerate(zip(degrees, titles)):
    plt.subplot(1, 3, i+1)

    # 다항식 회귀 모델
    poly_model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])

    poly_model.fit(X_demo, y_demo)
    y_pred_demo = poly_model.predict(X_test_demo)

    # 훈련 점수와 일반화 점수
    train_score = poly_model.score(X_demo, y_demo)
    test_score = r2_score(y_test_true, y_pred_demo)

    plt.scatter(X_demo, y_demo, alpha=0.6, label='Training Data')
    plt.plot(X_test_demo, y_test_true, 'g-', alpha=0.8, label='True Function')
    plt.plot(X_test_demo, y_pred_demo, 'r-', alpha=0.8, label='Prediction')
    plt.title(f'{title}\\nTrain R²: {train_score:.3f}, Test R²: {test_score:.3f}')
    plt.legend()
    plt.ylim(-2, 2)

plt.tight_layout()
plt.show()

print("\\n💡 해석:")
print("- 왼쪽 (차수=1): 너무 단순해서 패턴을 못 찾음 (과소적합)")
print("- 가운데 (차수=4): 적절한 복잡도로 실제 패턴 포착")
print("- 오른쪽 (차수=15): 너무 복잡해서 노이즈까지 학습 (과적합)")
````
